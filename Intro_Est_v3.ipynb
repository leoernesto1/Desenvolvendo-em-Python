# Introdução a Estatística

Bom pessoal, esses exercícios tem o objetivo demonstrar em Python como calcular as medidas mais comuns em estatística. Também exemplificaremos como fazer alguns gráficos,  que podem nos ajudar a entender melhor nossos dados. 


Primeiro importamos as bibliotecas que iremos usar nos exercícios.

```python
import numpy as np
import pandas as pd

```

Agora vamos criar uma simples lista com alguns valores, que usaremos no nosso primeiro exemplo.

```python
dta = pd.Series([899,89,6,11,65,1,16,11,89,1,165,8,99,11,6,89,11,89,56,511,6,11,8,9,9,116,5,6,1165,11,89,899,5,1,165,
      8,9,9,1165,6,5,11,89,11,89,5,5,5,6,1,165,11,89,5,11,89,5,89,5,6,11,89,11,89,5,11,89,5,11,89,5,11,89,5,
     11,89,5,11,8,99,11,89,5,11,89,5,6,5,89,6,5,6,5,899,1165])
```

Podemos ordenar os dados, para isso usamos sorted()

```python
dta_ord = sorted(dta)
print(dta_ord)
```

Podemos contar quantos valores temos no nosso exemplo, usaremos para isso o len()

```python
len(dta)
```

Como podemos somar todos os dados de uma lista? 
Simples, podemos usar o sum().


```python
aux = sum(dta)
aux
```

Já que temos a somatória dos valores e a quantidade de valores, podemos calcular a média da seguinte maneira: 

```python
md = aux/95
md

#ou também
#md = aux/len(dta)
```

O Numpy também nos oferece uma maneira mais facil de calcular a média, usando np.mean()


```python
aux = np.mean(dta)
aux
```

Para o cálculo da mediana, teremos que desenvolver um pouco mais de código, usando condicionais.  

```python

if len(dta_ord) % 2 == 0:   
    
    aux2 = len(dta_ord)//2

    mediana = (dta_ord[aux2 + 1] + dta_ord[aux2])//2
                                                         

else:
    
    mediana = dta_ord[round((len(dta_ord)/2))]
    
    
mediana
```

Podemos calcular a mediana também usando np.median() de maneira mais simples, e não é necessário ordenar os dados previamente.

```python
print(np.median(dta))
print(np.median(dta_ord))
```

Vamos agora calcular a variância populacional e o desvio padrão, para isso usaremos o For: 

```python
aux3 = 0
for i in dta:
      aux3 +=  ((i - md)**2)
        
var = aux3//len(dta)
print(var)


dp = np.sqrt(var)
print(dp)
```

Agora calcularemos usando np.var() e np.std(): 

```python
print(np.var(dta))
print(np.std(dta))
```

Iremos demonstar como calcular o coeficiente de variação:

```python
coef_var = dp/md *100
coef_var
```

Boxplot ou gráfico de caixa, é uma ferramenta gráfica usada para mostrar a variação de uma variável numérica, por meio de quartis. A linha no meio da caixa é o valor da mediana também chamado de segundo quartil, abaixo da mediana fica o primeiro quartil , acima o terceiro quartil. A caixa tem limites inferiores e superiores, os dados fora desse limite são chamados de Outliers.

```python
sns.set(style="darkgrid", color_codes= True)
sns.boxplot(data=dta_flores)
```


Gráfico de dispersão ou Scatter Plot, é usado para observar a distribuição de atributos par a par.


```python
X = dta_flores['sepal_length']
Y = dta_flores['sepal_width']
plt.scatter(X,Y)
plt.show()
```
